FUNCIONES DE UTILIDAD PARA LIMPIEZA DATASET

```python

def clean_and_convert(value):
    if isinstance(value, str):
        try:
            return int(value.split()[0])
        except ValueError:
            return 0  
    return value


# Función para contar los valores nulos por columna
def count_nulls(df):
    return df.isnull().sum()

# Función para eliminar filas con valores nulos
def drop_nulls(df, inplace=False):
    if inplace:
        df.dropna(inplace=True)
    else:
        return df.dropna()

# Función para eliminar columnas específicas
def drop_columns(df, columns, inplace=False):
    if inplace:
        df.drop(columns=columns, inplace=True)
    else:
        return df.drop(columns=columns)

# Función para eliminar filas duplicadas
def drop_duplicates(df, inplace=False):
    if inplace:
        df.drop_duplicates(inplace=True)
    else:
        return df.drop_duplicates()

# Función para convertir una columna a tipo float y devolver valores incorrectos
def convert_to_float(df, column, inplace=False):
    try:
        converted = pd.to_numeric(df[column], errors='coerce').astype(float)
        if inplace:
            df[column] = converted
        else:
            return converted
    except ValueError as e:
        print(f"Error en la conversión de '{column}' a tipo float: {e}")

# Función para convertir una columna a tipo entero y devolver valores incorrectos
def convert_to_int(df, column, inplace=False):
    try:
        converted = pd.to_numeric(df[column], errors='coerce').astype(pd.Int64Dtype())
        if inplace:
            df[column] = converted
        else:
            return converted
    except ValueError as e:
        print(f"Error en la conversión de '{column}' a tipo entero: {e}")

```
